```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
news = pd.read_csv("news.csv", index_col=False)
```

```python
list = news.head(1)
```

```python
for i in news.text[1]:
    print(i,end='')
```

```python
for i in news.title[1]:
    print(i,end='')
```

## If you see the two outputs above and try to gauge the feature importance from the column name, Text is the most important feature here

## Let's try to use these features and then categorize them into REAL or FAKE news

## first we have to use TF-IDF vectorizer to get the TF and the IDF of the text column!!

## Let us split the data into train and test parts

```python
X = news['text']
Y = news['label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1234)
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_df=0.7)
```

## the above command initializes the tfidf vectorizer with the stop words being in English and the maximum document frequency being 0.7

## max_df is used to remove the words that occur very frequently in the corpus, basically these type of words are called, "corpus specific stop-words", here max_df = 0.7, ignore the stop words that are repeated in less than 70% of the documents.

## min_df is used to remove the words that do not occur frequently in the corpus, if we use min_df = 0.01, the words that repeat in less than 1% of the documents

```python
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
for i in X_train_tfidf[0]:
    print(i)
```

```python
for i in tfidf.get_feature_names_out():
    print(i)
```

## the above the cells depict the matrix for the tfidf vectorizer and the subsequent cell shows the feature names of the transformation

```python
X_train_tfidf.shape
```

```python
X_test_tfidf.shape
```

## the shapes after tfidf vectorization, these features are passed to the PasssiveAggressive Classifier

## PassiveAggressiveClassifier is a online machine learning technique, where the classifier is has mini-batch kind of updation when the data is fed to the machine learning algorithm. The simplest way to understand the behaviour of the algo is that it acts passive for the correct classification and acts aggressively for the wrong classifications. This kind of behaviour is needed for this kind of dataset where we have to classify if the news is REAL or FAKE!

```python
PAC = PassiveAggressiveClassifier()
```

```python
PAC.fit(X_train_tfidf, Y_train)
```

```python
y_pred = PAC.predict(X_test_tfidf)
```

```python
score = accuracy_score(Y_test, y_pred)
```

```python
print(score)
```
